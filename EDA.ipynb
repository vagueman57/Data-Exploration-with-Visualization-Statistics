# **6. EDA using Bivariate and Multivariate Analysis**

```python
import pandas as pd
import seaborn as sns
```

```python
titanic = pd.read_csv('/content/Titanic-Dataset.csv')
titanic.head()
```

```python
tips = pd.read_csv('/content/tips.csv')
tips.head()
```

```python
flight = pd.read_csv('/content/routes.csv')
flight.head()
```

```python
iris = pd.read_csv('/content/Iris.csv')
iris.head()
```

###**1. ScatterPlot(Numerical-Numerical)**

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker', size='size')
```

###**2. BarPlot (Numerical-categorical)**

```python
titanic.head()
```

```python
sns.barplot(x='Pclass', y='Age', data=titanic, hue='Sex')
```

```python
sns.barplot(x='Pclass', y='Fare', data=titanic, hue='Sex')
```

###**3. Box Plot(Numerical -Categorical)**

```python
sns.boxplot(x='Sex', y='Age', data=titanic, hue='Survived')
```

###**4. DisPlot (Numerical-Categorical)**

```python
sns.displot(titanic[titanic['Survived']==0]['Age'])
```

```python
sns.displot(data=titanic, x='Age', hue='Survived', kind='kde')
```

###**5. HeatMap(categorical-categorical)**

```python
titanic.head()
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='pie')
```

```python
titanic.groupby('Embarked')['Survived'].mean()*100
```

```python
titanic.groupby('Sex')['Embarked'].value_counts(normalize=True) * 100
```

```python
titanic.groupby('Pclass')['Embarked'].value_counts(normalize=True) * 100
```

### **6. ClusterMap(categorical-categorical)**

```python
sns.heatmap(pd.crosstab(titanic['SibSp'], titanic['Survived']))
```

```python
sns.pairplot(iris, hue='Species')
```
